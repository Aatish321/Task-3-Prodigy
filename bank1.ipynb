```python
bk1 = bk.copy()
bk1.columns
```

```python
# Select only numeric columns
numeric_df = bk1.select_dtypes(include=[np.number])

# Compute the correlation matrix for numeric columns
corr_matrix = numeric_df.corr().abs()

# Proceed with identifying highly correlated features
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

high_corr_cols = [column for column in upper_tri.columns if any(upper_tri[column] > 0.9)]
bk1.drop(high_corr_cols, inplace=True, axis=1)
print(bk1.columns)



```

```python
# Get the list of columns that are in both the DataFrame and high_corr_cols
columns_to_drop = [col for col in high_corr_cols if col in bk1.columns]

# Drop only those columns
bk1.drop(columns_to_drop, inplace=True, axis=1)

# Display the remaining columns
print(bk1.columns)
```

```python
bk1.shape
```

```python
from sklearn.preprocessing import LabelEncoder

# Create an instance of LabelEncoder
lb = LabelEncoder()

# Apply LabelEncoder to each column in the DataFrame
bk_encoded = bk1.apply(lb.fit_transform)

# Display the encoded DataFrame
bk_encoded
```

```python
bk_encoded['deposit'].value_counts()
```

```python
x = bk_encoded.drop('deposit',axis =1)
y = bk_encoded['deposit']
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

```python
from sklearn.model_selection import train_test_split
print(4119*0.25)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size =0.25, random_state =1)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

def eval_mode(y_test, y_pred):
    acc = accuracy_score(y_test, y_pred)
    print('Accuracy Score:', acc)
    cm = confusion_matrix(y_test, y_pred)
    print('Confusion Matrix:\n', cm)
    print('Classification Report:\n', classification_report(y_test, y_pred))


def mscore(model):
    train_score = model.score(x_train, y_train)
    test_score = model.score(x_test, y_test)
    print('Training Score:', train_score)
    print('Testing Score:', test_score)
  
```

```python
from sklearn.tree import DecisionTreeClassifier

bk = DecisionTreeClassifier(criterion ='gini', max_depth =5, min_samples_split =10)
bk.fit(x_train, y_train)
```

```python
mscore(bk)
```

```python
ypred_bk = bk.predict(x_test)
print(ypred_bk)
```

```python
eval_mode(y_test, ypred_bk)
```
